##### Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

ANS:
Here are five common plots you can create using Seaborn along with their typical use cases
i)`scatterplot`
ii)`displot`
iii)`relplot`
iv)`catplot`
v)`jointplot`

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.scatterplot(x = tips.total_bill, y = tips.tip)
```

```python
sns.displot(tips['tip'])
```

```python
sns.relplot(x = tips.total_bill, y = tips.tip, data = tips ,style = 'size', hue = 'smoker')
```

```python
sns.catplot(x = 'day', y= 'total_bill',data = tips)
```

```python
sns.jointplot(x = tips.total_bill, y = tips.tip, data = tips,hue = 'time')
```

#### Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x ='pclass', y = 'age' and y = 'fare'.

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic
```

```python

plt.figure(figsize=(12,10))
plt.subplot(2,2,1)
sns.boxplot(x = 'pclass', y = 'age',data = titanic)
plt.title('age by passenger class')
plt.subplot(2,2,2)
sns.boxplot(x = 'pclass', y = 'fare',data = titanic)
plt.title('fare by passenger class')
```
